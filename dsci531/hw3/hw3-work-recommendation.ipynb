## Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

countries = ['England', 'China', 'India', 'Mexico', 'Ethiopia']
pronouns = ['he', 'she']

def standardize_job_title(title):
    """Standardize similar job titles for consistent analysis"""
    title_lower = title.lower()
    if any(term in title_lower for term in ['software engineer', 'software developer', 'web developer', 'developer']):
        return 'Software Developer/Engineer'
    elif 'data' in title_lower and any(term in title_lower for term in ['analyst', 'scientist']):
        return 'Data Analyst/Scientist'
    elif any(term in title_lower for term in ['customer service', 'customer support']):
        return 'Customer Service Rep'
    elif any(term in title_lower for term in ['sales', 'account executive']):
        return 'Sales Representative'
    elif any(term in title_lower for term in ['freelance writer', 'content writer', 'copywriter']):
        return 'Writer/Content Creator'
    elif any(term in title_lower for term in ['administrative assistant', 'virtual assistant']):
        return 'Administrative Assistant'
    elif any(term in title_lower for term in ['delivery driver', 'driver']):
        return 'Delivery Driver'
    elif any(term in title_lower for term in ['data entry', 'data entry clerk']):
        return 'Data Entry Clerk'
    elif any(term in title_lower for term in ['retail', 'retail associate', 'sales associate']):
        return 'Retail Associate'
    elif any(term in title_lower for term in ['warehouse', 'warehouse worker']):
        return 'Warehouse Worker'
    elif any(term in title_lower for term in ['digital marketing', 'marketing']):
        return 'Marketing Specialist'
    elif any(term in title_lower for term in ['tutor', 'teacher', 'esl teacher']):
        return 'Tutor/Teacher'
    elif any(term in title_lower for term in ['project manager']):
        return 'Project Manager'
    else:
        return title
```

```python
processed_main_data = {}
main_job_probabilities = {}

for key, job_lists in main_prompt_data.items():
    gender, country = key
    standardized_jobs = []
    for job_list in job_lists:
        standardized_jobs.append([standardize_job_title(job) for job in job_list])
    
    processed_main_data[key] = standardized_jobs
    
    job_counts = Counter()
    for job_list in standardized_jobs:
        job_counts.update(set(job_list))
    
    total_runs = len(job_lists)
    main_job_probabilities[key] = {job: (count / total_runs * 100) for job, count in job_counts.items()}

processed_baseline_data = {}
baseline_job_probabilities = {}

for gender, job_lists in baseline_prompt_data.items():
    standardized_run_jobs = []
    
    for job_list in job_lists:
        standardized_jobs = []
        
        for job_entry in job_list:
            if "/" in job_entry and not any(x in job_entry for x in ["Developer/Engineer", "Analyst/Scientist", "Support/Help"]):
                split_jobs = [standardize_job_title(j.strip()) for j in job_entry.split("/")]
                standardized_jobs.extend(split_jobs)
            elif "(" in job_entry and ")" in job_entry:
                job_base = standardize_job_title(job_entry.split("(")[0].strip())
                standardized_jobs.append(job_base)
            else:
                standardized_jobs.append(standardize_job_title(job_entry))
        
        standardized_run_jobs.append(standardized_jobs)
    
    processed_baseline_data[gender] = standardized_run_jobs
    
    job_counts = Counter()
    for job_list in standardized_run_jobs:
        job_counts.update(set(job_list))
    
    total_runs = len(job_lists)
    baseline_job_probabilities[gender] = {job: (count / total_runs * 100) for job, count in job_counts.items()}

main_job_counts = main_job_probabilities
baseline_job_counts = baseline_job_probabilities

all_jobs = []
for key, job_lists in processed_main_data.items():
    for job_list in job_lists:
        all_jobs.extend(job_list)

job_counter = Counter(all_jobs)
top_jobs = [job for job, count in job_counter.most_common(10)]

plot_data = []
for job in top_jobs:
    for country in countries:
        for gender in pronouns:
            main_key = (gender, country)
            if main_key in main_job_counts and job in main_job_counts[main_key]:
                main_percentage = main_job_counts[main_key][job]
            else:
                main_percentage = 0
            
            if gender in baseline_job_counts and job in baseline_job_counts[gender]:
                base_percentage = baseline_job_counts[gender][job]
            else:
                base_percentage = 0
            
            plot_data.append({
                'Job': job,
                'Country': country,
                'Gender': gender,
                'Main Percentage': main_percentage,
                'Baseline Percentage': base_percentage,
                'Bias': main_percentage - base_percentage
            })
df_plot = pd.DataFrame(plot_data)

plt.figure(figsize=(15, 8))
colors = sns.color_palette('husl', len(countries))
country_color_map = dict(zip(countries, colors))
job_positions = {job: i for i, job in enumerate(top_jobs)}
for job in top_jobs:
    job_data = df_plot[df_plot['Job'] == job]
    
    for idx, row in job_data.iterrows():
        country = row['Country']
        gender = row['Gender']
        x_pos = job_positions[job]
        y_main = row['Main Percentage']
        y_base = row['Baseline Percentage']
        
        jitter = np.random.uniform(-0.2, 0.2)
        color = country_color_map[country]
        marker = 'o' if gender == 'he' else '^'
        plt.scatter(x_pos + jitter, y_main, color=color, marker=marker, s=80, alpha=0.7,
                   label=f"{country} ({gender})" if job == top_jobs[0] else "")

plt.xlabel('Job Type')
plt.ylabel('Probability of Recommendation (%)')
plt.title('Job Recommendation Probability by Country and Job Type')
plt.grid(True, alpha=0.3, axis='y')
plt.xticks(range(len(top_jobs)), top_jobs, rotation=45, ha='right')
plt.ylim(0, 70)

handles, labels = plt.gca().get_legend_handles_labels()
plt.ylim(-10, 110)
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), loc='upper right', title='Country (Gender)')

plt.tight_layout()
plt.savefig('Probability: Job Recommendation.png', dpi=300)
plt.show()
```

```python
bias_summary = df_plot.groupby(['Country', 'Job'])['Bias'].mean().reset_index()
significant_biases = bias_summary[abs(bias_summary['Bias']) > 10].sort_values('Bias', ascending=False)

print("Most significant biases in job recommendations by country and job type:")
print(significant_biases)

heatmap_data = df_plot.pivot_table(
    values='Main Percentage', 
    index='Country',
    columns='Job',
    aggfunc='mean'
)

plt.figure(figsize=(15, 8))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", fmt=".1f", linewidths=.5)
plt.title('Heatmap: Job-Country')
plt.tight_layout()
plt.savefig('country_job_heatmap.png', dpi=300)
plt.show()
```
